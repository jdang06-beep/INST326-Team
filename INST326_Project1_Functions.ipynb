```python
#Eliab Getachew
#Making a function that allows you to leave notes on movies to rewatch them in the future if  they are worth a rewatch
# list for storing movies and notes
movies = []
#making a function that allows  you to add notes to a movie
def add_note():
    title = input("Movie name: ")
    watch_again = input("Would you watch this again? (y/n): ")
    watch_again = watch_again.lower()
    user_note = input("Write your note: ")
    # add everything to the list
    movies.append([title, watch_again, user_note])
    print("\nNote saved!")
def see_rewatch():
    print("\n--- Movies to Rewatch ---")
    count = 0
    # go through all movies
    for i in range(len(movies)):
        if movies[i][1] == "y":
            print("\nMovie:", movies[i][0])
            print("Note:", movies[i][2])
            count = count + 1
    if count == 0:
        print("No movies to rewatch")
def see_all():
    print("\n--- All Notes ---")

    if len(movies) == 0:
        print("No notes saved")
    else:
        for i in range(len(movies)):
            print("\nMovie:", movies[i][0])
            if movies[i][1] == "y":
                print("Rewatch: yes")
            else:
                print("Rewatch: no")
            print("Note:", movies[i][2])
# main part
running = True
while running == True:
    print("\n1 - Add note")
    print("2 - See rewatch list")
    print("3 - See all notes")
    print("4 - Exit")
    pick = input("\nWhat do you want to do? ")
    if pick == "1":
        add_note()
    elif pick == "2":
        see_rewatch()
    elif pick == "3":
        see_all()
    elif pick == "4":
        print("bye")
        running = False
    else:
        print("pick 1 2 3 or 4")
```

```python
#Eliab
#Scored review based out of 10 saves it and gives a review after of all the movies you entered and informs the user to
#pick a better movie or to watch this movie if it's good.

# List to store (movie, rating) pairs
movie_list = []
# defining a function that gives a reviw based on rating from user
def review_movie(name, rating):
    if rating < 0 or rating > 10:
        print("rating should be between 0-10.")
    elif rating < 4:
        print(f"{name} scored {rating}/10. \nThis movie is horrible DO NOT WATCH!.")
    elif rating < 6:
        print(f"{name} scored {rating}/10. \nIt's okay maybe try a different movie.")
    elif rating < 8:
        print(f"{name} scored {rating}/10. \nIt's good worth watching.")
    elif rating < 10:
        print(f"{name} scored {rating}/10. \nIt's great you should definitely watch it.")
    else:
        print(f"{name} scored {rating}/10. \nIt's a phenomenal movie a must watch!")
while True:
    # Asks for movie name from user
    movie_name = input("\nEnter a movie name or type 'quit' to stop: ").strip()
    if movie_name.lower().strip() == "quit":
        break  # ends loop
    # Ask for rating
    try:
        movie_rating = float(input("Rate the movie out of 10: ").strip())
    except ValueError:
        print("Please enter a number for the rating (e.g., 7.5).")
        continue
    # Give review and recommendation
    review_movie(movie_name, movie_rating)
    # Save the movie and its rating
    movie_list.append((movie_name, movie_rating))
    # Print all saved movies so far
    print("\nMovies you've rated so far:")
    for movie, rating in movie_list:
        print(f"- {movie}: {rating}/10")
# Final summary
print("\nThanks for rating movies! Here's your final list:")
for movie, rating in movie_list:
    print(f"- {movie}: {rating}/10")
```

```python
# Jolina
reviews = [
    {"title": "The Conjuring", "content": "Terrifying and well-directed.", "genre": "Horror", "rating": 4.7},
    {"title": "Mission Impossible", "content": "High energy and intense action!", "genre": "Action", "rating": 4.3},
    {"title": "Gone Girl", "content": "Dark, clever, and suspenseful.", "genre": "Thriller", "rating": 4.6},
    {"title": "Insidious", "content": "Creepy atmosphere and great scares.", "genre": "Horror", "rating": 4.5},
]


def categorize_reviews_by_genre(reviews):
# takes a list of review dictionaries and groups them by genre. each review should have a 'genre' key. returns a dictionary where each key is a different genre and the values are reviews.
  categorized_by_genre = {}

  for review in reviews:
    genre = review['genre'].lower()
    if genre not in categorized_by_genre:
      categorized_by_genre[genre] = []
    categorized_by_genre[genre].append(review)

  return categorized_by_genre

grouped_reviews = categorize_reviews_by_genre(reviews)
print(grouped_reviews)
```

```python
# Jolina
def categorize_reviews_by_rating(reviews):
# takes a list of review dictionaries and groups them by positive or negative. each review should have a 'positive' or 'negative' key. returns a dictionary where a key is 'positive' or 'negative' and the values are reviews.
  categorized_by_rating = {'positive': [], 'negative': []}

  for review in reviews:
    rating = review['rating']
    if 0 <= rating <= 2.5:
      categorized_by_rating['negative'].append(review)
    else:
      categorized_by_rating['positive'].append(review)

  return categorized_by_rating

grouped_reviews = categorize_reviews_by_rating(reviews)
print(grouped_reviews)
```

```python
# Jolina
def store_reviews_by_title(reviews):
  # takes a list of review dictionaries and returns a dictionary where the key is the media identifier and the value is the full review dictionary.
  reviews_by_title = {}

  for review in reviews:
    title = review['title']
    reviews_by_title[title] = review

  return reviews_by_title
```
